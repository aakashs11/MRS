```python
#replacing nan values with 0 in case of numbers and '' for others,removing extra characters from runtime and boxoffice and converting to int

df["Runtime"]=df["Runtime"].str.extract('(^\d*)')
#df['BoxOffice']=df['BoxOffice'].str.extract('(^\d*)')
#df['Imdbrating']=df['Imdbrating']
#df["Runtime"] = df["Runtime"].str.rstrip('hmin')
df["Runtime"].fillna('0',inplace=True)
#df['BoxOffice'].fillna('0',inplace=True)
df.fillna('',inplace=True)

df["Runtime"]=df["Runtime"].astype('int64')
df["Runtime"]=df["Runtime"].astype('int64')
```

```python
#df['Actors'].replace('',np.nan,inplace=True)
#df['Director']=df['Director'].replace('',np.nan,inplace=True)
#df['Director']=df['Director'].dropna()
#dropping unnamed columns
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df['Runtime']=df['Runtime'].replace(np.nan,df['Runtime'].mean(),regex=True)
df['Runtime'].value_counts()
df['Actors'].isnull().sum()
```

```python
#15 colums and 4848 movies
df=df.tail(-1)#removing first row its repeated
df.head(10)
df.shape
```

```python
# discarding the commas between the actors' full names and getting only the first three names
df['Actors'] =  [x.split(',')[:3] for x in df['Actors'] ]
df = df.fillna('missing value')
df['Genre'] = df['Genre'].map(lambda x: x.split(','))
# putting the genres in a list of words
#df['Genre'] = df['Genre'].map(lambda x: x.lower().split(','))
#df['Genre'] = [x.split(',') for x in df['Genre'] ]
df['Director'] = [x.split(',') for x in df['Director'] ]
```

```python
# merging together first and last name for each actor and director, so it's considered as one word 
# and there is no mix up between people sharing a first name
for index, row in df.iterrows():
    df['Actors'][index] = [x.lower().replace(' ','') for x in df['Actors'][index]]
    df['Director'][index] = [x.lower().replace(' ','') for x in df['Director'][index]]
    
```

```python
df1 = df[['Title','Genre','Director','Actors','Plot']]
df1.head()
```

```python
#keyword extraction
#Create a empty list Keywords
df1['keywords'] = ''
#Loop across all rows to extract all keywords from description
for index, row in df1.iterrows():
    Plot = row['Plot']
 #instantiating Rake by default it uses English stopwords from NLTK and discards all punctuation chars
    r = Rake()
     #extract words by passing the text
    r.extract_keywords_from_text(Plot)
    #get the dictionary with key words and their scores
    keyword_dict_scores = r.get_word_degrees()
    #assign keywords to new columns
    row['keywords'] = list(keyword_dict_scores.keys())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer #tokenizes a collection of words extracted from a text doc
from ast import literal_eval #This evaluates whether an expresion is a Python datatype or not
#count vectorizer
#igenerating the count matrix
#instantiating and generating the count matrix
count = CountVectorizer()
tfidf= TfidfVectorizer()
count_matrix = count.fit_transform(df1['bow'])
count_matrix_tfidf=tfidf.fit_transform(df1['bow'])
#create a Series for movie titles so they are associated to an ordered numerical list, we will use this later to match index
indices = pd.Series(df1.iloc[:,0],df1.index)
indices
```

```python
#generating the cosine similarity matrix

cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim_tfidf=cosine_similarity(count_matrix_tfidf,count_matrix_tfidf)
```

```python
def recommendations(title,n,cosine_sim = cosine_sim):
    recommended_movies = []
    recommended_movies_tfidf = []
    
    #get index of the movie that matches the title
    idx=indices[indices==title].index[0]
    
    #find highest cosine_sim this title shares with other titles extracted earlier and save it in a Series
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    score_series_tfidf=pd.Series(cosine_sim_tfidf[idx]).sort_values(ascending = False)
    #get indexes of the 'n' most similar movies
    top_n_indexes = list(score_series.iloc[1:n+1].index)
    top_n_indexes_tfidf = list(score_series_tfidf.iloc[1:n+1].index)
    print(top_n_indexes)
    
    #populating the list with titles of n matching movie
    for i in top_n_indexes:
        recommended_movies.append(list(indices)[i])
        recommended_movies_tfidf.append(list(indices)[i])
    return recommended_movies,recommended_movies_tfidf
```
